### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

# Promedio
suma = 0
for i, esperanza in enumerate(esperanzas_de_vida):
    suma += esperanza
promedio = suma / len(esperanzas_de_vida)

print(f"el promedio es: {promedio:.2f}")

# Máximo
mayor = esperanzas_de_vida[1]
for i, esperanza in enumerate(esperanzas_de_vida):
    if mayor < esperanza:
        mayor = esperanza

print(f"El valor mayor es: {mayor}")

# Mínimo
minimo = esperanzas_de_vida[1]
for i, esperanza in enumerate(esperanzas_de_vida):
    if minimo > esperanza:
        minimo = esperanza

print(f"El valor minimo es: {minimo}")

````

### 1.3 — Funciones

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí
def clasificar_esperanza_vida(años):
    if años>=75:
        return "Alta"
    elif años>=60 and años<=74.9:
        return "Media"
    else:
        return "baja"

for i in range(5):
    edad = float(input("Ingrese numero de años: "))
    esperanza = clasificar_esperanza_vida(edad)
    print(f"la esperanza de vida para los años de {edad} es: {esperanza}")
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí
df_nuevo = df.copy()

df_nuevo["esperanza_vida"] = df_nuevo["esperanza_vida"].fillna(df_nuevo["esperanza_vida"].mean())
df_nuevo["gasto_salud_usd"] = df_nuevo["gasto_salud_usd"].fillna(df_nuevo["gasto_salud_usd"].mean())
df_nuevo["mortalidad_infantil"] = df_nuevo["mortalidad_infantil"].fillna(df_nuevo["mortalidad_infantil"].mean())
df_nuevo["medicos_por_1000"] = df_nuevo["medicos_por_1000"].fillna(df_nuevo["medicos_por_1000"].mean())
df_nuevo["obesidad_pct"] = df_nuevo["obesidad_pct"].fillna(df_nuevo["obesidad_pct"].mean())
df_nuevo["diabetes_pct"] = df_nuevo["diabetes_pct"].fillna(df_nuevo["diabetes_pct"].mean())
df_nuevo["tabaquismo_pct"] = df_nuevo["tabaquismo_pct"].fillna(df_nuevo["tabaquismo_pct"].mean())
df_nuevo["camas_hospital_por_1000"] = df_nuevo["camas_hospital_por_1000"].fillna(df_nuevo["camas_hospital_por_1000"].mean())
df_nuevo["vacunacion_pct"] = df_nuevo["vacunacion_pct"].fillna(df_nuevo["vacunacion_pct"].mean())

df_nuevo.isnull().sum()
````

**Responde aquí** (doble clic para editar):

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

> Tu respuesta: primero porque usamos la media, por que es el valor dependiendo de la complejidad del dato es la mas acertada para poder seguir haciendo el proceso estadistico e incluso es la que mas se aproxima a el dato y existe una alta posibilidad de que si el dato faltante estuviera en el dataframe, este dato estaria muy cerca de la media. Y porque no las eliminamos, a mayor informacion mas relevancia y acertada es la informacion.
